# 4. vòng lặp

- `for` + `range(start, stop, end)`

```python
# vong lap while
i = 0
while i > 7:
    
    if i == 6:
        continue
    if i == 3:
        pass
    if i == 2:
        break #dung vong lap
    print(i)
    i += 1
    
```
